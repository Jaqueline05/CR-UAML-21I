```python
#UNIVERSIDAD AUTONOMA METROPOLITANA UNIDAD LERMA
#Programadora WENDY JAQUELINE HERNANDEZ ARMENTA
#Fecha: 19-04-2021
import networkx as nx #Importacion de librerias
import matplotlib.pyplot as plt
lista_enlaces = [("Alicia","Beto"), #Realizacion de la lista de enlaces
             ("Carlos","Alicia"),
             ("Alicia","David"),
             ("Alicia","Carmen"),
             ("Alicia","Monica"),
             ("Beto","Jorge"),
             ("Jorge","Enrique"),
             ("Enrique","Rosa"),
             ("Rosa","Jorge"),
             ("Enrique","Irene"),
             ("Irene","Jorge"),
             ("Irene","Rosa"),
             ("Carmen","Monica"),
             ("David","Carlos"),
             ("Carlos","Monica"),
             ("Carmen","Irene")]
G = nx.Graph(lista_enlaces) #se Crea la Red
nx.draw(G,with_labels=True,node_color="PINK") # Se dibuja la red
plt.savefig("AMIGOS.png")
```

```python
# nodos
nodos = len(nx.nodes(G))  #Numero de nodos 
# enlaces
Enlaces = len(nx.edges(G)) #Numero de enlaces 
# diametro
Diametro = nx.diameter(G)    #Diametro
print("Informacion de la red: \n- nodos = " + str(nodos) +
      "\n- enlaces = " + str(Enlaces) +
      "\n- diámetro = " + str(Diametro) )
nx.info(G)#se imprime la informacion de la red 
```

```python

def num_amigos(nodo):
    amigos = nx.degree(G,nodo)
    print(str(nodo) + " tiene " + str(amigos) + " amigos")
    return amigos

dist_grados = nx.degree_histogram(G) #calcula la distribución de grados de la red
print(dist_grados)
```

```python
plt.title("Distribucion de Grados")
plt.plot(dist_grados) 
plt.xlabel("grado")
plt.ylabel("# nodos")
plt.savefig("distribucion.png")
```

```python
kn = nx.complete_graph(nodos)# crea una gráfica completa con nodos,
nx.draw(kn,with_labels=True,node_color="BLUE")
#plt.savefig("Nodos.png")
```

```python

densidad = nx.density(G) # Densidad de la red
print("La densidad de la red es " + str(densidad))
```
