```python
#UNIVERSIDAD AUTONOMA METROPOLITANA UNIDAD LERMA 
#Programadora WENDY JAQUELINE HERNANDEZ ARMENTA
#FECHA 12-04-2021
import networkx as nx #Importacion de librerias
import matplotlib.pyplot as plt

lista_enlaces = [("Alicia","Beto"), #Realizacion de la lista de enlaces
             ("Carlos","Alicia"),
             ("Alicia","David"),
             ("Alicia","Carmen"),
             ("Alicia","Monica"),
             ("Beto","Jorge"),
             ("Jorge","Enrique"),
             ("Enrique","Rosa"),
             ("Rosa","Jorge"),
             ("Enrique","Irene"),
             ("Irene","Jorge"),
             ("Irene","Rosa"),
             ("Carmen","Monica"),
             ("David","Carlos"),
             ("Carlos","Monica"),
             ("Carmen","Irene")]
G = nx.Graph(lista_enlaces) #se Crea la Red
nx.draw(G,with_labels=True,node_color="pink") # Se dibuja la red

nodos = nx.nodes(G) #Devuelve los nodos 
nodos

enlaces = nx.edges(G) # Devuelve los enlaces 
enlaces

camino = nx.shortest_path(G,"Carlos","Irene")  # Calcula uno de los caminos mas cortos 
print("Camino más corto entre Carlos e Irene " + str(camino))

caminos = nx.all_shortest_paths(G,"Carlos","Enrique") #Calcula todos los caminos mas cortos 
for camino in caminos: print("El camino más corto entre Carlos y Enrique " + str(camino))
    
distancia = nx.shortest_path_length(G,"Carlos","Irene")#Calcula la longitud
print("La distancia del camino mas corto entre los nodos Carlos e Irene " + str(distancia))

diametro = nx.diameter(G) #Calcula el diametro de la red
print("El diámetero de la red es " + str(diametro))

prom_dist = nx.average_shortest_path_length(G) #Calcula la distancia promedio de la red
print("La distancia promedio es " + str(prom_dist))
```
